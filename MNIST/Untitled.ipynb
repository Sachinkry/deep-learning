```python
def percpetron(weight, bias, x):
    model = np.add(np.dot(x, weight), bias)
    logit = 1/(1+np.exp(-model))
    print('Type: {}'.format(logit))
    return np.round(logit)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Inputs and labels for an AND gate
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])  # Input
y = np.array([0, 0, 0, 1])  # Output

# Weights and bias for the AND gate (manually set)
w = np.array([1, 1])
b = -1.5

# Activation function (step function)
def activation(x):
    return np.where(x > 0, 1, 0)

# Compute the net input
z = np.dot(X, w) + b

# Apply the activation function
output = activation(z)

# Plot the inputs
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.winter)

# Create a mesh to plot the decision boundary
x_min, x_max = -0.5, 1.5
y_min, y_max = -0.5, 1.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

# Compute the net input for the mesh points
Z = activation(w[0] * xx + w[1] * yy + b)

# Plot the decision boundary
plt.contourf(xx, yy, Z, cmap=plt.cm.winter, alpha=0.2)

# Label axes and show plot
plt.xlabel('Input 1')
plt.ylabel('Input 2')
plt.title('Decision Boundary for AND Gate')
plt.show()
```
